```python
#Download csv file from github

import requests
url = 'https://raw.githubusercontent.com/codebasics/deep-learning-keras-tf-tutorial/master/6_gradient_descent/insurance_data.csv'
res = requests.get(url, allow_redirects=True)
with open('insurance_data.csv','wb') as file:
    file.write(res.content)
sales_team = pd.read_csv('insurance_data.csv')
```

```python
df = pd.read_csv('insurance_data.csv')
df.head()
```

```python
# Spilitting data into test and train:

from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(df[['age','affordibility']],df.bought_insurance,test_size = 0.2, random_state = 25)
```

```python
x_train
```

```python
# Feature Scaling:

x_train_scaled = x_train.copy()
x_train_scaled['age'] = x_train_scaled['age']/100

x_test_scaled = x_test.copy()
x_test_scaled['age'] = x_test_scaled['age']/100
```

```python
x_train_scaled
```

```python
from keras import Sequential
from keras.layers import Dense
```

```python
model = Sequential([
    Dense(1 , input_shape = (2,), activation = 'sigmoid' , kernel_initializer = 'ones', bias_initializer = 'zeros')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.fit(x_train_scaled , y_train , epochs = 5000)
```

```python
model.evaluate(x_test_scaled , y_test)
```

```python
model.predict(x_test_scaled)
```

```python
# To find the Final weights and bias:
w , b = model.get_weights()
print(w)
print(b)
```

```python
# Applying gradient descent by python:

def log_loss(y_true , y_predicted):
    
    epsilon = 1e-15
    y_predicted_new = [max(i,epsilon) for i in y_predicted]
    y_predicted_new = [min(i,epsilon) for i in y_predicted]
    y_predicted_new = np.array(y_predicted_new)
    # This was done so that we can get rid of zero by relacing it with a small number close to zero
    # as log(0) tends to infinity.
    
    out = -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
    return out
```

```python
def sigmoid_np(x):
    return 1/(1+np.exp(x))
sigmoid_np(np.array([1,2,3]))
```

```python
def gradient_descent(age , affordibility , y_true ,rate ,epochs):
    
    #initialise parameters:
    w0 = 0
    w1 = 0
    b = 0
    n = len(age)
    
    for i in range(epochs):
        #Calculating 
        weighted_sum = w0*age + w1*affordibility + b
        y_predicted = sigmoid_np(weighted_sum)
        
        #Calculating loss Function:
        loss = log_loss(y_true , y_predicted)
        
        #calculating Derivatives:
        dj_dw0 = (1/n)*np.dot(np.transpose(age),(y_predicted - y_true))
        dj_dw1 = (1/n)*np.dot(np.transpose(affordibility),(y_predicted - y_true))
        dj_db = np.mean(y_predicted-y_true)
        
        #Applying gradient descent:
        w0 = w0 - rate*dj_dw0
        w1 = w1 - rate*dj_dw1
        b = b - rate*dj_db
        
        print(f'Epoch:{i} , w0: {w0} , w1: {w1} , loss: {loss}')
    return w0,w1,b
```
